In England the currency is made up of pound, £, and pence, p, and there are eight coins in general circulation:

    1p, 2p, 5p, 10p, 20p, 50p, £1 (100p) and £2 (200p).

It is possible to make £2 in the following way:

    1×£1 + 1×50p + 2×20p + 1×5p + 1×2p + 3×1p

How many different ways can £2 be made using any number of coins?

```python
matches=0
for n1 in range(0,201):
    for n2 in range(0, 101):
        for n5 in range(0,41):
            for n10 in range(0,21):
                for n20 in range(0,11):
                    for n50 in range(0,5):
                        for n100 in range(0,3):
                            for n200 in range(0,2):
                                if(n1*1+n2*2+n5*5+n10*10+n20*20+n50*50+n100*100+n200*200)==200:
                                    matches=matches+1
    print("Iteration: ",n1)
print("Solution: ", matches)                
```

USING A £1 COIN:

```python
matches=0
for n1 in range(0,101):
    for n2 in range(0, 51):
        for n5 in range(0,21):
            for n10 in range(0,11):
                for n20 in range(0,6):
                    for n50 in range(0,3):
                        if(n1*1+n2*2+n5*5+n10*10+n20*20+n50*50)==100:
                            matches=matches+1
print("Solution: ", matches)    
```
